```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.load('../results/bs4096K10LR0.1E10D0.001DR0.4.npz')
plt.plot([i for i in range(1, len(results['tr_loss']) + 1)], results['tr_loss'], label = 'train')
plt.plot([i for i in range(1, len(results['t_loss']) + 1)], results['t_loss'], label = 'test')
plt.gca().set_ylim(bottom=0)
plt.title('losses per epoch w/o regularization')
plt.legend()
plt.show()
```

```python
r1 = np.load('../results/loss_bs8192K5LR0.01E15D0.001DR0.4ALPH0.01.npz')['t_loss']
r2 = np.load('../results/loss_bs8192K5LR0.01E15D0.001DR0.4ALPH1.npz')['t_loss']
r3 = np.load('../results/loss_bs8192K5LR0.01E15D0.001DR0.4ALPH0.0001.npz')['t_loss']
r4 = np.load('../results/loss_bs8192K5LR0.01E15D0.001DR0.4ALPH0.npz')['t_loss']


plt.plot([i for i in range(1, len(r1) + 1)], r1, label = 'Alpha = 0.01')
plt.plot([i for i in range(1, len(r1) + 1)], r3, label = 'Alpha = 0.0001')
plt.plot([i for i in range(1, len(r2) + 1)], r2, label = 'Alpha = 1')
plt.plot([i for i in range(1, len(r4) + 1)], r4, label = 'Alpha = 0')
plt.gca().set_ylim(bottom=0)
plt.title('losses by alpha')
plt.legend()
plt.show()
```

```python
r1 = np.load('../results/bs64K10LR0.01E25D1e-05.npz')['t_loss']
r2 = np.load('../results/bs64K10LR0.001E25D1e-05.npz')['t_loss']
r3 = np.load('../results/bs64K10LR0.0001E25D1e-05.npz')['t_loss']

plt.plot([i for i in range(1, len(r1) + 1)], r1, label = 'lr=0.01')
plt.plot([i for i in range(1, len(r2) + 1)], r2, label = 'lr=0.001')
plt.plot([i for i in range(1, len(r1) + 1)], r3, label = 'lr=0.0001')
plt.gca().set_ylim(bottom=0)
plt.title('losses by learning rate')
plt.legend()
plt.show()
```
